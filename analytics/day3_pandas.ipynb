```python
# =======================================
# THE DATAFRAME -  The Heart of Pandas
# ======================================
# A Dataframe is a table with: 
# > ROWS - eah row is one record/observation
# > COLUMNS - each column is one variable/feature
# > INDEX - the row number (starts at 0)
# 
# Think of it exactly like an Excel spreadsheet
# but one that can hold millions of rows and 
# process them in second s
# 
# We crate a DataFrame from a dictionary 
# Remeber dictionaries from Day 1?
# keys = colum names 
# values = the data in each column 

# creating the first DataFrame 
server_data = {
    "server_name": ["SVR01", "SVR02", "SVR03",
                    "SVR04", "SVR05"],
    "cpu":         [45, 89, 23, 67, 91],
    "ram":         [60, 78, 34, 55, 88],
    "disk":        [70, 85, 45, 60, 92],
    "location":    ["JHB", "CPT", "JHB", 
                    "DBN", "JHB"],
    "status":      ["OK", "WARNING", "OK", 
                    "OK", "CRITICAL"]
}

# pd.DataFrame() converts the dictionary to a table 
df = pd.DataFrame(server_data)

print("--- MY FIRST DATADRAME ----")
print(df)
print()

# THE 5 MOST IMPORTANT COMMANDS
# run these on EVERY new dataset i ever work with 
print("--- THE 5 ESSENTIAL COMMANDS ----")
print()

# 1. shape - how many rows and colums 
print(f"1. Shape: {df.shape}")
print(f" {df.shape[0]} rows, {df.shape[1]} columns")
print()

# 2. dtpes - what data type is each column 
print("2. Data Types:")
print(df.dtypes)
print()

# 3. head() - first 5 rows
# Always check the top of your data first
print("3. First 3 rows (head):")
print(df.head(3))
print()

# 4. describe() - statistics for number columns 
# count, mean, min , max , percentiles 
print("4. Statistics (describe):")
print(df.describe().round(1))
print()

# 5. isnull().sum() - count missing values 
# Missing data break ML models 
# Always check this before doing anything 
print("5. Missing values per column:")
print(df.isnull().sum())

    
```

```python
# ===========================================
# ACCESSING DATA -  getting Rows and Columns
# ===========================================
# There are two ways to access data in a DataFrame
# 
# df['column_name'] - get ONE column 
# df[['col1' , 'col2']] - get multiple columns 
# 
# df.loc[row] - access by LABEL/NAME
# df.iloc[row] - access by POSITION (like Numpy)
#
# Think of it like this: 
# df['cpu'] = "give me the CPU column"
# df.iloc[0] = give me row at position 0"
# df.loc[0] = "give me row with index label0" 
# in our case they are the same because our 
# index is 0,1,2,3,4, by default 

print("--- ACCESSING COLUMNS ---")
print()

# get ONE column returns a Series 
# A series is like a single column Dataframe
# It has an index on the left and values on the right
cpu_column = df['cpu']
print("CPU column only:")
print(cpu_column)
print(f"Type: {type(cpu_column)}")
print()

# get MULTIPLE columns - returns a DataFrames
# Notice the DOUBLE square brackets [[ ]]
# Outer brackets = pandas selection
# Inner brackets = the list of columns
resources = df[['server_name', 'cpu', 'ram', 'disk']]
print("Server resources only:")
print(resources)
print()


print("--- ACCESSING ROWS ---")
print()

# iloc - access by INTERGER position 
# iloc[0] = first row (position 0)
# iloc[-1] = last row
# iloc{0:3} = first 3 rows 
print("First row (iloc[0]):")
print(df.iloc[0])
print()

print("Last row (iloc[-1]):")
print(df.iloc[-1])
print()

print("First 2 rows (iloc[0:2]):")
print(df.iloc[0:2])
print()

print("---- ACCESSING SPECIFIC CELLS ---")
print()

# Get ONE specific value
# iloc[row_position, column_position]
cpu_svr02 = df.iloc[1, 1] # row 1, column 1 = SVR-02 CPU
print(f"SVR-02 CPU reading: {cpu_svr02}%")

# loc[row_label, column_name] - more readable 
cpu_svr02_loc = df.loc[1, 'cpu']
print(f"Same value using loc: {cpu_svr02_loc}%")

print()
print("--- COLUMN STATISTICS ---")
print()

# You can run statistics on any column directly 
print(f"Average CPU: {df['cpu'].mean():.1f}%")
print(f"Highest CPU: {df['cpu'].max()}%")
print(f"Lowest CPu: {df['cpu'].min()}%")
print(f"Which server has highest CPu:")
# idxmax() returns the INDEX of the maxium value
max_idx = df['cpu'].idxmax()
print(f" Index {max_idx}: {df.loc[max_idx, 'server_name']}")
print(f" CPU: {df.loc[max_idx, 'cpu']}%")
print(f" Status: {df.loc[max_idx, 'status']}")







```

```python
# ================================================
# FILTERING DATA - Findintg What you You Need
# ===============================================
# Filtering means selecting only rows that meet
# a condition - exactly like SQL WHERE clause
# or Excel AutoFilter
# 
# SYNTAX: df[df['column'] condition value]
#
# Read it like this:
# df[ ] = "from my DataFrame, give me rows where..."
# df['cpu'] > 70 = "cpu column is greater than 70
#
# This returns a NEW DataFrame with ony
# the matching rows - orignal df is unchanged

print("--- BASIC FILTERING ---")
print()

# Filter 1: Servers with CPU above 70%
high_cpu = df[df['cpu'] > 70]
print("Servers with CPU above 70:%")
print(high_cpu)
print()

# Filter 2: Servers with CRITICAL Status
critical = df[df['status'] == "CRITICAL"]
print("CRITICAL server:")
print(critical)
print()

# Filter 3 Servers in JHB only
jhb_servers = df[df['location']== 'JHB']
print("JHB server only:")
print(jhb_servers)
print()

print("--- MULTIPLE CONDITIONS ---")
print()

# AND condition - both must be True 
# Use & symbol (not 'amd)
# Important: wrap each condition in brackets ()
high_cpu_jhb = df[(df['cpu'] > 70) & 
             (df['location'] == 'JHB')]
print("High CPU AND in JHB:")
print(high_cpu_jhb)
print()

# OR condtion - either can be True 
# Use | symbol (not 'or')
warning_or_critical = df[(df['status'] == 'WARNING') | 
                        (df['status'] == 'CRITICAL')]
print("WARNING or CRITICAL servers:")
print(warning_or_critical)
print()

# NOT  condtion - reverse the filter 
# Use - symbol (not 'not')
not_ok = df[~(df['status'] == 'OK')]
print("Servers that are NOT OK:")
print(not_ok)
print()

print("--- PRACTICAL FILTERS ---")
print()

# isin() - check if value is in a list
# Like SQL: WHERE location IN ('JHB', 'CPT')
major_cities = df[df['location'].isin(['JHB', 'CPT'])]
print("Servers in JHB or CPT:")
print(major_cities)
print()

# between() - filter a range of values
medium_cpu = df[df['cpu'].between(40, 80)]
print("Servers with CPU between 40% and 80%:")
print(medium_cpu)
print()

# How many servers need attention?
needs_attention = df[df['cpu'] > 70]
print(f"Servers needing attention: "
      f"{len(needs_attention)} of {len(df)}")
print(f"Percentage: "
      f"{len(needs_attention)/len(df)*100:.0f}%")
      

      
```

```python
# =============================================================
# ADDING AND MODIFYING COLUMNS 
# =============================================================
# DataFrames are set not static - you can add new
# columns, change existing ones, calculate 
# new values fromexisting columns
# 
# This is how you ENRICH your data 
# Raw data rarely has everything you need
# You can calculate new colums from what you have 
#
# SYNTAX: df['new_column'] = calculation

print("--- ADDING NEW COLUMNS ---")
print()

# Add a column that combines cpu + ram + disk 
# into a single health score
# Highern score = more loaded = worse health
df['load_score'] = (df['cpu'] + 
                    df['ram'] +
                    df['disk']) / 3
print("Load score added (average of cpu+ram+disk):")
print(df[['server_name', 'cpu',
          'ram', 'disk', 'load_score']])
print()

# Add an alert level column based on cpu
# This uses np.where()- like an if/else for columns
# np.where(condition, value_if_true, value_if_false)
df['alert'] = np.where(df['cpu'] > 85,
                       'RED',
              np.where(df['cpu'] > 65,
                       'AMBER',
                       'GREEN'))
print("Alert levels added:")
print(df[['server_name', 'cpu', 'alert']])
print()

# Add a boolean column - is the server healthy?
df['is_healthy'] = df['cpu'] < 70

print("Is healthy column added:")
print(df[['server_name', 'cpu', 'is_healthy']])
print()

print("--- MODIFYING EXISTING COLUMNS ---")
print()

# Make server names uppercase
# .str. lets you apply string operations to a column 
df['server_name'] = df['server_name'].str.upper()
print("Server names uppercased:")
print(df['server_name'].tolist())
print()

# Round load_score to 1 decimal place
df['load_score'] = df['load_score'].round(1)
print("Load scores rounded:")
print(df[['server_name', 'load_score']])
print()

print("--- FINAL ENRICHED DATAFRAME ---")
print()
print(df)
print()

# Summary of alerts
print("--- ALERT SUMMARY ---")
print(df['alert'].value_counts())
print()
healthy_count = df['is_healthy'].sum()
print(f"Healthy servers: {healthy_count} of {len(df)}")
print(f"Unhealthy servers: {len(df) - healthy_count} of {len(df)}")
    
```

```python
# =====================================================
# GROUPBY - Analysing Data by Category 
# =====================================================
# groupby() groups your data by a column 
# then lets you calculate statistics PER GROUP 
#
# Think of it like this
# "Give me the average CPU per LOCATION"
# "Give me the count of servers per STATUS"
# "Give me the max disk per CITY"
# 
# It is exacly like SQL GROUP BY
# SELECT location, AVG(cpu) FROM servers
# GROUP BY location
# 
# SYNTAX
# df.groupby('column)['other_column'].function()
# 
# Step 1: groupby('location) split into groups 
# Step 2: ['cpu'] - pick which column to analyse
# Step 3: .mean() - what to calculate per group

print("--- GROUPBY BASICS ---")
print()

# Average CPU per location
avg_cpu_location = df.groupby('location')['cpu'].mean()
print("Average CPU per location:")
print(avg_cpu_location.round(1))
print()

# Count servers per location
count_per_location = df.groupby('location')['server_name'].count()
print("Number of servers per location:")
print(count_per_location)
print()

# Max CPU per location
max_cpu_loaction = df.groupby('location')['cpu'].max()
print("Highest CPU per location:")
print(max_cpu_loaction)
print()

print("--- GROUPBY WITH MULTIPLE CALCULATIONS ---")
print()

# agg() lets you calculate MULTIPLE statistics 
# at once for the same group 
# Pass a dictionary: column -> list of functions 
location_summary = df.groupby('location').agg(
    server_count = ('server_name', 'count'),
    avg_cpu      = ('cpu', 'mean'),
    max_cpu      = ('cpu', 'max'),
    avg_ram      = ('ram', 'mean'),
    avg_load     = ('load_score', 'mean')
).round(1)

print("Full location summary:")
print(location_summary)
print()

# groupby status
status_summary = df.groupby('status').agg(
    count         = ('server_name', 'count'),
    avg_cpu       = ('cpu', 'mean'),
    avg_load      = ('load_score', 'mean')
).round(1)

print("Summary by status:")
print(status_summary)
print()

print("--- GROUPBY WITH SORTHING ---")
print()

# Sort results to find worst location first 
worst_locations = df.groupby('location')['cpu'].mean()\
                    .sort_values(ascending=False)\
                    .round(1)
print("Locations ranked by average CPU (worst first):")
for location, avg in worst_locations.items():
    if avg > 70:
        flag = " X HIGH"
    elif avg > 50:
        flag = " > MEDIUM"
    else: 
        flag = " - LOW"
    print(f" {location}: {avg}% {flag}")
print()
print("--- GROUP INSIGHT ----")
print()

# Which location has the most unhealthy servers?
unhealthy =df[df['is_healthy'] == False]
unhealthy_by_location = unhealthy.groupby(
    'location')['server_name'].count()

print("Unhelthy servers per location:")
print(unhealthy_by_location)
print()
print(f"Location needing most attention:"
      f"{unhealthy_by_location.idxmax()}")
    
    
    
```
