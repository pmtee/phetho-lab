```python
# ================================================
# CONCEPT 1: VARIABLES
# ================================================
# A variable is a labelled box that stores a value
# name = "Phetho" means:
#   - create a box
#   - label it 'name'
#   - put 'Phetho' inside
# When Python sees 'name' later it opens
# the box and uses what is inside

name     = "Phetho"
surname  = "Tlaka"
city     = "Midrand"
country  = "South Africa"

print(f"Name:    {name} {surname}")
print(f"City:    {city}")
print(f"Country: {country}")
```

```python
# ================================================
# CONCEPT 2: THE 4 DATA TYPES
# ================================================

# TYPE 1 — String: any text, always in quotes
first_name  = "Phetho"
last_name   = "Tlaka"
full_name   = first_name + " " + last_name
print("--- STRING ---")
print(f"Full name:   {full_name}")
print(f"Characters:  {len(full_name)}")
print(f"Uppercase:   {full_name.upper()}")
print(f"Lowercase:   {full_name.lower()}")

# TYPE 2 — Integer: whole numbers, no decimals
print("\n--- INTEGER ---")
age          = 23
year_started = 2026
months_goal  = 12
print(f"Age:              {age}")
print(f"Year started:     {year_started}")
print(f"Months to goal:   {months_goal}")
print(f"Goal year:        {year_started + 1}")

# TYPE 3 — Float: decimal numbers
print("\n--- FLOAT ---")
hours_per_day  = 4.0
days_per_week  = 5.0
weeks          = 26.0
total_hours    = hours_per_day * days_per_week * weeks
print(f"Hours per day:       {hours_per_day}")
print(f"Days per week:       {days_per_week}")
print(f"Weeks in 6 months:   {weeks:.0f}")
print(f"Total study hours:   {total_hours:,.0f}")
print(f"Full working days:   {total_hours / 8:.0f}")

# TYPE 4 — Boolean: only True or False
print("\n--- BOOLEAN ---")
is_learning     = True
has_github      = True
has_python_cert = False
has_vm_setup    = True
print(f"Is learning:        {is_learning}")
print(f"GitHub created:     {has_github}")
print(f"Has Python cert:    {has_python_cert}")
print(f"VM set up:          {has_vm_setup}")
```

```python

# ================================================
# CONCEPT 3: LISTS
# ================================================
# A list stores multiple values in order
# Uses square brackets [ ]
# Each item has a position called an INDEX
# IMPORTANT: counting starts at 0 not 1

skills_learning = [
    "Python",
    "Docker",
    "Kubernetes",
    "AWS",
    "Power BI",
    "Machine Learning",
    "DevSecOps"
]

print("--- MY LEARNING LIST ---")
print(f"Full list:       {skills_learning}")
print(f"Total skills:    {len(skills_learning)}")
print(f"First skill:     {skills_learning[0]}")   # index 0 = first
print(f"Last skill:      {skills_learning[-1]}")  # -1 = always last
print(f"First 3:         {skills_learning[0:3]}") # slicing

# Add and remove items
skills_learning.append("Linux")
print(f"\nAfter adding Linux: {skills_learning}")
print(f"Now {len(skills_learning)} skills to learn")
```

```python
# ================================================
# CONCEPT 4: DICTIONARIES
# ================================================
# A dictionary stores key:value pairs
# Like a real dictionary — word (key) : meaning (value)
# Uses curly brackets { }

my_profile = {
    "name":        "Phetho Tlaka",
    "city":        "Midrand",
    "degree":      "BSc Computer Sciences",
    "university":  "Varsity College",
    "year":        2024,
    "github":      "github.com/pmtee",
    "goal":        "DevOps / ML Engineer"
}

print("--- MY PROFILE ---")
for key, value in my_profile.items():
    print(f"  {key:12}: {value}")

print(f"\nAccessing one value:")
print(f"My goal is to become a: {my_profile['goal']}")
```

```python
# ================================================
# CONCEPT 5: IF / ELSE — Making Decisions
# ================================================
# if checks a condition
# elif means 'else if' — check another condition
# else runs if nothing above matched
# IMPORTANT: indentation (spaces) is required

study_hours_today = 4

print("--- STUDY HOURS CHECKER ---")

if study_hours_today >= 5:
    message = "Outstanding! You went above target."
elif study_hours_today >= 3:
    message = "Good work. You hit your minimum target."
elif study_hours_today >= 1:
    message = "You showed up. Tomorrow do more."
else:
    message = "You missed today. Do not miss tomorrow."

print(f"Hours studied: {study_hours_today}")
print(f"Feedback: {message}")

# Comparing values
print("\n--- COMPARISON OPERATORS ---")
print(f"5 > 3  : {5 > 3}")    # greater than
print(f"5 < 3  : {5 < 3}")    # less than
print(f"5 == 5 : {5 == 5}")   # equal to (TWO equals signs)
print(f"5 != 3 : {5 != 3}")   # not equal to
print(f"5 >= 5 : {5 >= 5}")   # greater than or equal
```

```python
# ================================================
# CONCEPT 6: FOR LOOP
# ================================================
# A for loop repeats code for every item in a list
# It stops automatically when the list ends

roadmap = [
    "Month 1: Python + Data Analytics + Power BI",
    "Month 2: Machine Learning + Docker + AWS",
    "Month 3: LLMs + AI Agents + Kubernetes",
    "Month 4: C# + DevSecOps + CI/CD pipelines",
    "Month 5: MLOps + SageMaker + Terraform",
    "Month 6: Capstone project + Job applications"
]

print("--- 6 MONTH ROADMAP ---")
for i, month in enumerate(roadmap, 1):
    print(f"  {i}. {month}")

# Loop with a range
print("\n--- COUNTING WITH RANGE ---")
print("Days of study this week:")
for day in range(1, 6):
    print(f"  Day {day} of 5")
```

```python
# ================================================
# CONCEPT 7: FUNCTIONS
# ================================================
# A function is reusable code that does one job
# def = define the function
# Parameters = inputs (in brackets)
# return = the output

def greet(name, goal):
    """
    This is a docstring — it explains what the function does.
    This function greets a person and states their goal.
    """
    message = f"Hello, {name}. You goal is: {goal}."
    return message

def study_progress(days_completed, total_days):
    """Calculate and display study progress."""
    percentage = (days_completed / total_days) * 100
    remaining  = total_days - days_completed
    return {
        "completed":   days_completed,
        "remaining":   remaining,
        "percentage":  round(percentage, 1)
    }

# Call the functions
print("--- FUNCTION RESULTS ---")
result = greet("Phetho", "DevOps / ML Engineer")
print(result)

progress = study_progress(1, 180)
print(f"\nProgress after Day 1:")
print(f"  Days completed:  {progress['completed']}")
print(f"  Days remaining:  {progress['remaining']}")
print(f"  Progress:        {progress['percentage']}%")
print(f"  Keep going.")
```
