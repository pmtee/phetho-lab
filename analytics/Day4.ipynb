```python
# ==================================================================
# DAY 4- Working With Real Data 
# Phetho Tlaka | April 2026
# Dataset: Titanic passangers (891 real records)
# =================================================================

import pandas as pd
import numpy as np


# Load the CSV file into a DataFrame
df = pd.read_csv('titanic.csv')

print("Titanic dataset loaded!")
print()
print("=" * 45)
print(" STEP 1: UNDERSTAND YOUR DATA")
print("=" * 45)
print()
print(f"Shape: {df.shape}")
print(f" {df.shape[0]} passengers (rows)")
print(f" {df.shape[1]} columns of information")
print()
print("Column names and data types:")
print(df.dtypes)
print()
print("First 5 rows:")
print(df.head())


```

```python
# =================================================
# STEP2: CHECK DATA QUALITY 
# =================================================
# Real data is ALWAYS messy
# Before any analysis you must understand
# 1. What data is missing 
# 2. What does each column mean? 
# 3. Are the dat types correct?
#
# Nan = Not a Number = missing value 
# Pandas use Nan to represent msssing data 
# Missing data can break your analysis 
# You must decide: fill it or drop it

print("=" * 45)
print(" STEP 2: CHECK DATA QUALITY")
print("=" * 45)
print()

# isnull() returns True for missing values
# .sum() counts the Trues per column
# .sort_values() sort highest to lowest
missing = df.isnull().sum().sort_values(ascending=False)
print("Missing values per column:")
print(missing)
print()

# Show missing as percentage
missing_pct = (df.isnull().sum() / len(df) * 100)\
              .sort_values(ascending=False)
print("Missing values as percentage:")
for col, pct in missing_pct.items():
    if pct > 0:
        bar = "█"  * int(pct / 5)
        print(f" {col:12}: {pct:5.1f}% {bar}")
print()

# What do our columns mean?
print("=" * 45)
print(" COLUMN DESCRIPTIONS")
print("=" * 45)
columns_explained = { 
    "PassangerId": "Unique ID for each passenger",
    "Survived":    "0 = Dies, 1 = Survived",
    "Pclass":       "Ticket class: 1=First, 2=Second, 3=Third",
    "Name":        "Full name of passenger",
    "Sex":         "male or female",
    "Age":         "Age in years (177 missing)",
    "SibSp":       "Siblings/spouses abroad",
    "Parch":       "Parents/childern abroad",
    "Ticket":      "Ticket number",
    "Fare":        "Ticket price in pounds",
    "Cabin":       "Cabin number (687 missing)",
    "Embarked":    "Port: s=Southhampton c=Cherbourg Q=Queenstown"
}

for col, desc in columns_explained.items():
    print(f" {col:12}: {desc}")
print()

# Basic statistics
print("=" * 45)
print(" BASIC STATISTICS")
print("=" * 45)
print(df.describe().round(2))

    

```

```python
# ================================================================
# STEP 3: CLEAN THE DATA 
# ================================================================
# Cleaning means fixing mssing values and 
# making data consistent and usable
# 
# 3 strategies for mssing data:
# 
# 1. DROP - remove rows/columns with mssing data 
#
# 1. DROP - remove rows/columns with missing data 
#  Use when: too much data missing to be useful 
# Example: Cabin is 77% missing - drop the column
#
# 2. FILL WITH MEDIAN - replace missing with median 
# Use when: ttoo much dat missing to be useful
# Example:Cabin is 77% missing - drop the column
# 
# 2. FILL IN WITH MEDIAN - replace missing with median
# Use when: numerical column, not too many missing 
# Example: Age 20% missing - fill with median age
# WHY MEDIAN not MEAN?
# Mean is affected by outliers (very old/young passengers)
# Median is the middle value - more repreentative
#
# 3. FILL WITH MODE - replace with most common value
# Use when: text/category column
# Example: Embarked has 2 missing - fill with most common port

# Make a copy so we keep the orginal untouched 
# Always work on a copy - never destory the orginal data 
print("=" * 45)
print(" STEP 3: CLEAN THE DATA")
print("=" * 45)
print()

# Make a copy so we keep the orginal untouched 
# Always work on a copy - never destroy orginal data 
df_clean = df.copy()

print("Before cleaning")
print(f" Shape: {df_clean.shape}")
print(f" Missing Age:    {df_clean['Age'].isnull().sum()}")
print(f" Missing Cabin:  {df_clean['Cabin'].isnull().sum()}")
print(f" Missing Embarked: {df_clean['Embarked'].isnull().sum()}")
print()

# FIX 1: Drop Cabin column - 77% missing, not useful 
df_clean = df_clean.drop(columns=['Cabin'])
print("Fix 1: Dropped Cabin column (77% missing)")


# FIX 2: Fill missing Age with median age 
median_age = df_clean['Age'].median()
df_clean['Age'] = df_clean['Age'].fillna(median_age)
print(f"Fix 2: Filled missing Age with median ({median_age})")

# FIX 3: Fill missing Embarked with mode (most common)
most_common_port = df_clean['Embarked'].mode()[0]
df_clean['Embarked'] = df_clean['Embarked'].fillna(
                           most_common_port)
print(f"Fix 3: Filled missing Embarked with "
      f"mode ({most_common_port})")
print()


print("After cleaning:")
print(f" Shape: {df_clean.shape}")
print(f" Total missing values: " 
      f"{df_clean.isnull().sum().sum()}")
print()

# Verfiy no missintg vales remain 
print(" Missing values after cleaning:")
print(df_clean.isnull().sum())
print()

# Additional cleaning - make text consistent
#Converrt Sex to lowercase just in case
df_clean['Sex'] = df_clean['Sex'].str.lower()


# Convert Survived to clear labels 
# map() replaces values using a dictionary
df_clean['Survived_label'] = df_clean['Survived'].map(
    {0: 'Died', 1: 'Survived'}
)

print("Clean dataset preview:")
print(df_clean[['Name', 'Sex', 'Age', 
                'Pclass', 'Survived_label']].head(10))
      
```

```python
# ==================================================================
# STEP 4: ANALYSE THE DATA - Find The Insights
# ===================================================================
# Now that data is clean we can ask real questions
# This what data analysts do every day
# Questions drive analysis - not the other way around 
# 
# QUESTIONS WE WILL ANSWER:
# 1. What was teh overall survival rate?
# 2. Did gender affect survival?
# 3. Did ticket class affect survival?
# 4. Did age affect survival
# 5. Which port the best survival?


print("=" * 50)
print(" STEP 4: TITANIC DTA ANALYSIS")
print("=" * 50)
print()

# QUESTION 1: Overall survival rate 
total = len(df_clean)
survived = df_clean['Survived'].sum()
died = total - survived
survival_rate = survived / total * 100

print("---- Q1: OVERALL SURVIVAL ---")
print(f" Total passengers: {total}")
print(f" Survived:         {survived} "
      f"({100-survival_rate:.1f}%)")
print()

# QUESTION 2 Survival by gender
print("--- Q2: SURVIVAL BY GENDER ---")
gender_survival = df_clean.groupby('Sex')['Survived']\
                          .mean() * 100
for gender, rate in gender_survival.items():
    bar = "█" * int(rate /5)
    print(f" {gender:6}: {rate:.1f}% {bar}")

print()
print(" Insight: Women were far more likely")
print(" to urvive than men")
print()

# QUESTION 3: Survival by class
print("--- Q3: SURVIVAL BY TICKET CLASS ---")
class_survival = df_clean.groupby('Pclass')['Survived']\
                         .mean() * 100
class_names = {1: '1st Class',
               2: '2nd Class',
               3: '3rd Class'}
for pclass, rate in class_survival.items():
    bar = "█" * int(rate /5)
    print(f" {class_names[pclass]}: {rate:.1f}% {bar}")
print()
print(" Insight: First class passengers had")
print(" much better survival rates")
print()

# QUESTION 4 by age group
print("--- Q4: SURVIVAL BY AGE GROUP ---")

# pd.cut() divides a continous column into groups 
# bins = the boundaries of each group 
# labels = the name for each group 
df_clean['Age_group'] = pd.cut(
    df_clean['Age'],
    bins=[0, 12, 17, 35, 60, 100],
    labels=['Child', 'Teen',
            'Young Adult', 'Adult', 'Senior']
)

age_survival = df_clean.groupby(
    'Age_group', observed=True)['Survived'].mean() * 100
for age_group, rate in age_survival.items():
    bar = "█" * int(rate / 5)
    print(f" {str(age_group):12}: {rate:.1f}% {bar}")
print()
print(" Insight: Childern had the highest survival rate")
print()

# QUESTION 5: Survival by port
print("--- Q5: SURVIVAL BY PORT ----")
port_names = {
    'S': 'Southampton',
    'C': 'Cherbourg',
    'Q': 'Queenstown'
}

port_survival = df_clean.groupby(
    'Embarked')['Survived'].mean() * 100

for port, rate in port_survival.items():
    bar = "█" * int(rate /5)
    print(f" {port_names[port]:12}: {rate:.1f}% {bar}")
print()

# COMBINED INSIGHT
print("=" * 50)
print(" KEY INSIGHT SUMMARY")
print("=" * 50)
print(f" Overall survival:  {survival_rate:.1f}%")
print(f" Best gender:  "
      f"Female ({gender_survival['female']:1f}%)")
print(f" Besy class:    "
      f"1st ({class_survival[1]:.1f}%)")
print(f" Best age group: Children")
print(f" Best port: "
      f"Chebourg ({port_survival['C']:.1f}%)")
print("=" * 50)



 
```
