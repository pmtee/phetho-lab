```python
# =====================================================
# STEP 1: PREPARE DATA FOR MACHINE LEARNING 
# =====================================================
# ml MODELS CANNOT WORK WITH:
# - mISSING VALUES (NaN)
# - text data (male/female, S/C/Q)
# - Irrelevent columns (Nmae, Ticket)
#
# We must convert everything to NUMBERS
# This is called FEATURE ENGINEERING
# 
# FEATURES = the input columns the model uses 
#            to make predictions (X)
# TARGET   = the column we want to predict (y)
#            In our case: Survived (0 or 1)

# Load and clean data 
df = pd.read_csv('../analytics/titanic.csv')
df_clean = df.copy()
df_clean = df_clean.drop(columns=['Cabin'])
df_clean['Age'] = df_clean['Age'].fillna(
                  df_clean['Age'].median())
df_clean['Embarked'] = df_clean['Embarked'].fillna('S')

print("=" * 50)
print(" STEP 1: FEATURE ENGINNEERING")
print("=" * 50)
print()

# Convert sex to numbers 
# male =0 , female - 1
# This is called LABEL ENCODING 
df_clean['Sex_num'] = df_clean['Sex'].map(
    {'male':0, 'female': 1}
)
print("Sex encoded: male=0 , female=1")
print(df_clean[['Sex', 'Sex_num']].head(3))
print()

# Convert Embarked to numbers
# s=0, C=1,  Q=3
df_clean['Embarked_num'] = df_clean['Embarked'].map(
    {'S':0, 'C':1, 'Q': 2}
)
print("Embarked encoded: S=0, C=1, Q=2")
print(df_clean[['Embarked', 
                'Embarked_num']].head(3))
print()

# Select FEATURES - columns the models learns from
# We drop Name, Ticket - not useful for prediction
# We drop orginal text columns - use encoded ones
features = ['Pclass', 'Sex_num', 'Age',
            'SibSp', 'Parch', 'Fare',
            'Embarked_num']

x = df_clean[features] # Features (inputs)
y = df_clean['Survived'] # Target (what to predict)

print("=" * 50)
print("  FEATURES SELECTED")
print("=" * 50)
print(f"Feature matrix X shape: {X.shape}")
print(f" {X.shape[0]} passengers")
print(f" {X.shape[1]} features per passenger")
print()
print(f" Target y shape: {y.shape}")
print(f" 0 = Died:       {(y==0).sum()} passengers")
print(f" 1 = Survived:   {(y==1).sum()} passengers")
print()
print("Features we are using:")
for i, feature in enumerate(features, 1):
    print(f" {i}. {feature}")
```

```python
#  ==============================================================
# STEP 2: SPLIT DATA INTO TRAINING AND TEST SETS 
# ===============================================================
# We cannot train and test on the same dta 
# That would be like giving a student the answers 
# before the exam - of course they score 100%
# but they learned nothing
#
# Instead we split our data into two grups
# 
# TRAINING SET (80%) - model learns fro this 
# TEST SET (20%)     - we test accuracy on this
#                      model has NEVER seen this
# 
# This gives us an honest accuracy score 
# 
# train_test_split() does this automatically
# random_state=42 means the split is reproducible
# same seed = same split every time you run it

print("=" * 50)
print("  STEP 2: TRAIN/TEST SPLIT")
print("=" * 50)
print()

X_train, X_test, y_train, y_test = train_test_split(
    X,           # features
    y,           # traget
    test_size=0.2, # 20% for testing
    random_state=42 #reproducible split
)

print("Data split complete!")
print()
print(f"Training set: {X_train.shape[0]} passengers")
print(f" Model LEARNS from these")
print()
print(f" Test set:  {X_test.shape[0]} passengers")
print(f" Model is TESTED on these")
print(f" Model has NEVER seen these before")
print()

# Check the split balanced 
train_survival = y_train.mean() * 100
test_survival = y_test.mean() * 100

print(f" Training survival rate: {train_survival:.1f}%")
print(f"Test survival rate:      {test_survival:.1f}%")
print(f" Both close to overall 38.4% - good split!")
print()
print("=" * 50)
print("  ANALOGY")
print("=" * 50)
print(" Training set = textbook + practise questions")
print(" Test set =  final exam")
print(" Acurracy = your exam score")
print(" Good model = scores well on exam it")
print("             has never seen these before")





                       
```

```python
# ================================================================
# STEP 3: TRAIN THE MODEL 
# =================================================================
# We use Random Forest - one of the most powerful 
# and reliable ML algorithms ever created 
# 
# WHAT IS A RANDOM FOREST ?
# Imagine asking 100 different people 
# "Will this passengers survive?"
# Each person looks at different clues
# The mahority vote wins
# 
# Each "person" is a DECISION TREE
# A deciosion tree asks questions like:
# IS the passenger female?
#    YES > Survived
#    NO  > Died
#  NO > Was passenger in 1st Class?
#    YES > Survived 
#    NO > Died
#
# Random Forest builds 100 of these trees
# Each tree sees slightlhy different data 
# Final prediction = majority of all trees
# More treees = more accurate = more stable 

print("=" * 50)
print(" STEP 3: TRAIN THE MODEL")
print("=" * 50)
print()

# Create the model 
# n_estimators=100 means 100 decions trees 
# random_state =42 make resluts reproducible
model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

print("Model created: Random Forest")
print(f"  Trees: 100")
print(f"  Algorithm: Majority vote of 100 trees")
print()
print("Training started...")

# .fit() is where learning happens
# We pass X_train (features) and y_train (answers)
# The model studies all 712 passengers
# and learns which features predict survival
model.fit(X_train, y_train)

print("Training complete!")
print()
print("=" * 50)
print("  WHAT JUST HAPPENED?")
print("=" * 50)
print("  1. Model looked at 712 passengers")
print("  2. Built 100 decision trees")
print("  3. Each tree learned different patterns")
print("  4. Together they vote on predictions")
print()

# Feature importance — which features matter most?
# The model tells us which features were most
# useful for making predictions
importances = pd.Series(
    model.feature_importances_,
    index=features
).sort_values(ascending=False)

print("=" * 50)
print("  FEATURE IMPORTANCE")
print("  (Which clues matter most?)")
print("=" * 50)
for feature, importance in importances.items():
    bar = "█" * int(importance * 50)
    print(f"  {feature:12}: {importance:.3f} {bar}")
print()
print("  Higher = more important for prediction")
```

```python
# ================================================
# STEP 4: TEST THE MODEL — How Accurate Is It?
# ================================================
# Now we test our trained model on the 179
# passengers it has NEVER seen before
#
# .predict() uses the model to make predictions
# We compare predictions vs actual outcomes
#
# ACCURACY = correct predictions / total predictions
#
# Example:
# Actual:    [1, 0, 1, 1, 0]
# Predicted: [1, 0, 0, 1, 0]
# Correct:      ✓  ✓  ✗  ✓  ✓  = 4/5 = 80% accuracy

print("=" * 50)
print("  STEP 4: TEST THE MODEL")
print("=" * 50)
print()

# Make predictions on test set
# model has never seen X_test before
y_pred = model.predict(X_test)

print("Predictions made on 179 test passengers")
print()

# Compare first 10 predictions vs actual
print("First 10 predictions vs actual:")
print(f"  {'Actual':10} {'Predicted':10} {'Correct?':10}")
print(f"  {'-'*35}")
for actual, predicted in zip(
        y_test.values[:10], y_pred[:10]):
    actual_label = 'Survived' if actual == 1 else 'Died'
    pred_label = 'Survived' if predicted == 1 else 'Died'
    correct = '✓' if actual == predicted else '✗'
    print(f"  {actual_label:10} {pred_label:10} {correct}")
print()

# Overall accuracy
accuracy = accuracy_score(y_test, y_pred)
print("=" * 50)
print("  MODEL ACCURACY")
print("=" * 50)
print(f"  Correct predictions: "
      f"{int(accuracy * len(y_test))} of {len(y_test)}")
print(f"  Accuracy score: {accuracy * 100:.1f}%")
print()

if accuracy > 0.85:
    grade = "EXCELLENT"
elif accuracy > 0.80:
    grade = "VERY GOOD"
elif accuracy > 0.75:
    grade = "GOOD"
else:
    grade = "DECENT — room to improve"

print(f"  Grade: {grade}")
print()

# Detailed breakdown
print("=" * 50)
print("  DETAILED REPORT")
print("=" * 50)
print(classification_report(
    y_test, y_pred,
    target_names=['Died', 'Survived']
))

print("=" * 50)
print("  WHAT THESE NUMBERS MEAN")
print("=" * 50)
print("  Precision: when model says 'Survived'")
print("             how often is it right?")
print("  Recall:    of all who actually survived")
print("             how many did model catch?")
print("  F1-score:  balance of precision & recall")
print("  Support:   how many in each class")
```
